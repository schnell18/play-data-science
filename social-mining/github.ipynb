```python
from pytrends.request import TrendReq
import numpy as np
import pandas as pd
import configparser
import json
```

```python
keyword1 = 'New Zealand'
keyword2 = 'Australia'
pytrends = TrendReq(hl='en-US', tz=360)
pytrends.build_payload(kw_list=[keyword1, keyword2], cat=0, timeframe='today 5-y')
df = pytrends.interest_over_time()
```

```python
df.plot()
```

```python
%%sql
select v1.state,
       v1.year,
       round(v1.population / v2.area, 2)
  from state_areas.csv v2,
       (select 
               column0 as state,
               year,
               sum(population) as population
          from state_population.csv t1
         inner join state_abbrevs.csv t2 on (t1.state = t2.column1)
         where t1.state not in ('PR', 'USA')
         group by t2.column0, year
        ) v1
 where v1.state = v2.state
 order by 3 desc
    
```

```python
%%sql
select t1.state,
       t2.column0 as state,
       t1.year,
       t1.population
  from state_population.csv t1
  left join state_abbrevs.csv t2 on (t1.state = t2.column1)
 where t2.column0 is null
    
```

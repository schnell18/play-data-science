```python
import pandas as pd
import numpy as np
```

```python
# load student performance data
score = pd.read_csv('StudentsPerformance.csv')
score
```

# Math Operations
## Operations in columns

```python
# select a column and calculate total sum
score['math score'].average()
```

```python
# count, mean, std, max and min
score['math score'].count()
score['math score'].mean()
score['math score'].std()
score['math score'].max()
score['math score'].min()
```

```python
# use describe()
score.describe()
```

## Operations in rows

```python
# calculating the sum in a row
score['math score'] + score['reading score'] + score['writing score']
```

```python
# calculating average and assign to a new column 'avg score'
score['avg score'] = round((score['math score'] + score['reading score'] + score['writing score'])/3, 0)
```

```python
score
```

# value count

```python
# count by gender
score['gender'].value_counts()
```

```python
# relative frequency
score['gender'].value_counts(normalize=True)
```

```python
# counting "parental level of education"
score['parental level of education'].value_counts()
```

```python
score['parental level of education'].value_counts(normalize=True).round(2)
```

# Sort dataframe

```python
# sort by one column
score.sort_values(by='math score')
```

```python
# sort by one column descendingly
score.sort_values(by='math score', ascending=False)
```

```python
# sort by one column descendingly
score.sort_values(by='math score', ascending=False, inplace=True)
```

```python
# sort by multiple columns descendingly
score.sort_values(by=['math score', 'reading score'], ascending=False)
score
```

```python
# sort descednig with a key function
score.sort_values('race/ethnicity', ascending=True, key=lambda col:col.str.lower())
```

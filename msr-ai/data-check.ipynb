```python
import pandas as pd
```

```python
df_bq = pd.read_csv("contrib-bq.csv")
df_commit = pd.read_csv("commits-sum.csv", parse_dates=["author_date"])
```

```python
df_com_sum = df_commit.groupby("full_name", as_index=False).agg(contributors=pd.NamedAgg(column="author_name", aggfunc="nunique"))
```

```python
df_com_sum.sort_values("contributors", ascending=False)
```

```python
df_cmp = df_com_sum.merge(df_bq, how="left", left_on="full_name", right_on="repo")
df_cmp.query("repo == repo").sort_values("contributors_x", ascending=False)
```
